```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
from scipy import signal
```

```python
img=cv.imread('Downloads\leaf.jpg',0)
```

```python
x=np.arange(0,img.shape[1])
plt.plot(x,img[100])
plt.title("Intensity of 100th row of image")
```

```python
grad=np.gradient(img[100],x)
plt.plot(x,grad)
plt.title("Gradient of 100th row of image")
```

```python
retval=cv.getGaussianKernel(15,3)
xr=np.arange(-7,8)
xnew = np.linspace(xr.min(), xr.max(),300) 
spl = make_interp_spline(xr,retval,k=3) 
retval_smooth = spl(xnew)
plt.plot(xnew,retval_smooth)
plt.title("1D Gaussian kernal")
```

```python
blur=signal.fftconvolve(img[100],retval.squeeze(),mode='same')
plt.plot(x,blur)
plt.title("Noise removed 100th row")
```

```python
grad=np.gradient(blur,x)
plt.plot(x,grad)
plt.title("Grad of noise removed 100th row")
```

```python
kern_grad=np.gradient(retval.squeeze(),xr)
spl = make_interp_spline(xr,kern_grad,k=3) 
kernal_grad_smooth = spl(xnew)
plt.plot(xnew,kernal_grad_smooth)
plt.title("1D gaussian kernal grad")
```

```python
out=signal.fftconvolve(img[100],kern_grad.squeeze(),mode='same')
plt.plot(x,out)
plt.title("Conv between kernal grad and 100th row")
```

```python
retval_2=cv.getGaussianKernel(5,4)
retval_n=retval_2.reshape(1,-1)
retval_n=retval_n.squeeze()
filter2D=[]
for i in range(5):
    filter2D_r=[]
    s=[]
    for j in range(5):
         filter2D_r.append(retval_n[i]*retval_n[j])
    filter2D.append(filter2D_r)
filter2D=np.vstack(filter2D) 
xn=np.arange(-2,3)
filter2D_grad=[]
for k in range(5):
    filter2D_grad.append(np.gradient(filter2D[k],xn))
filter2D_grad=np.vstack(filter2D_grad)
filter2D_grad_tra=filter2D_grad.transpose()
fig,ax=plt.subplots(1,2,figsize=(9,9))
ax[0].imshow(cv.filter2D(img,-1,filter2D_grad),cmap='gray')
ax[1].imshow(cv.filter2D(img,-1,filter2D_grad_tra),cmap='gray')
ax[0].set_title("Vertical edges")
ax[1].set_title("Horizontal edges")
```

```python
ver_ed=cv.filter2D(img,-1,filter2D_grad)
hori_ed=cv.filter2D(img,-1,filter2D_grad_tra)
ver_ed_s=np.square(ver_ed)
hori_ed_s=np.square(hori_ed)
ed=np.sqrt(ver_ed_s+hori_ed_s)
ed=ed.astype('int')
plt.imshow(ed,cmap='gray')
plt.title("Edges")
```
