```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from skimage import measure as mea
```

```python
img=cv.imread('Downloads\lena.bmp',0)
```

```python
#partial derivatives of an image.In image x means rows and y means colomns 
edge_y=np.array([(-1,1)]).astype('float')
changes_respect_to_x=cv.filter2D(img,-1,edge_y) 
edge_x=np.array([(-1),(1)]).astype('float')
changes_respect_to_y=cv.filter2D(img,-1,edge_x)
```

```python
fig,ax=plt.subplots(1,2,figsize=(7,7))
ax[0].imshow(changes_respect_to_x,cmap='gray',vmin=-255,vmax=255)
ax[1].imshow(changes_respect_to_y,cmap='gray',vmin=-255,vmax=255)
ax[0].set_xticks([]),ax[0].set_yticks([])
ax[1].set_xticks([]),ax[1].set_yticks([])
ax[0].set_title("changes_respect_to_x")
ax[1].set_title("changes_respect_to_y")
plt.show()
```

```python
#sobel filter for edge detection
Sobel_vertical=np.array([(-1,0,1),(-2,0,2),(-1,0,1)]).astype('float')
Sobel_horizontal=np.array([(-1,-2,-1),(0,0,0),(1,2,1)]).astype('float')
vert_edge=cv.filter2D(img,-1,Sobel_vertical)
hori_edge=cv.filter2D(img,-1,Sobel_horizontal)
```

```python
fig,ax=plt.subplots(1,2,figsize=(7,7))
ax[0].imshow(vert_edge,cmap='gray',vmin=-255*4,vmax=255*4)
ax[1].imshow(hori_edge,cmap='gray',vmin=-255*4,vmax=255*4)
ax[0].set_xticks([]),ax[0].set_yticks([])
ax[1].set_xticks([]),ax[1].set_yticks([])
ax[0].set_title("Vertical edges")
ax[1].set_title("Horizontal edges")
plt.show()
```

```python
#prewitt fliter for edge detection
prewitt_vertical=np.array([(-1,0,1),(-1,0,1),(-1,0,1)]).astype('float')
prewitt_horizontal=np.array([(-1,-1,-1),(0,0,0),(1,1,1)]).astype('float')
vert_edge_p=cv.filter2D(img,-1,Sobel_vertical)
hori_edge_p=cv.filter2D(img,-1,Sobel_horizontal)
```

```python
fig,ax=plt.subplots(1,2,figsize=(7,7))
ax[0].imshow(vert_edge_p,cmap='gray',vmin=-255*3,vmax=255*3)
ax[1].imshow(hori_edge_p,cmap='gray',vmin=-255*3,vmax=255*3)
ax[0].set_xticks([]),ax[0].set_yticks([])
ax[1].set_xticks([]),ax[1].set_yticks([])
ax[0].set_title("Vertical edges")
ax[1].set_title("Horizontal edges")
plt.show()
```

```python
#intensity graph of the imagex
s=mea.profile_line(img,(0, 0),(0,img.shape[1]-1),linewidth=(img.shape[0]*2)-1,reduce_func=np.sum)
s_=s/512
x=np.arange(1,513)
plt.plot(x,s_)
plt.ylim(0,255)
plt.title("Intensity graph of image")
plt.show()
```

```python
#1st,2nd & 3rd derivative wrt x using Sobal kernal
h=[1,2,3] 
titles=["1st derivative_x","2nd derivative_x","3rd derivative_x"]
fig,ax=plt.subplots(1,3,figsize=(10,10))
for i in range (3):
    ax[i].imshow(cv.Sobel(img,-1,h[i],0,ksize=5),cmap='gray')#we can use the color map with vmin=-18*255 and vmax=18*255 but 
    #since edges are not that strong they won't be visible clearly
    ax[i].set_xticks([]),ax[i].set_yticks([])
    ax[i].set_title(titles[i])
plt.show()
```

```python
#1st,2nd & 3rd derivative wrt y using Sobal kernal
h=[1,2,3] 
titles=["1st derivative_y","2nd derivative_y","3rd derivative_y"]
fig,ax=plt.subplots(1,3,figsize=(10,10))
for i in range (3):
    ax[i].imshow(cv.Sobel(img,-1,0,h[i],ksize=5),cmap='gray')
    ax[i].set_xticks([]),ax[i].set_yticks([])
    ax[i].set_title(titles[i])
plt.show()
```

```python
#1st,2nd & 3rd derivative wrt xy using Sobal kernal
h=[1,2,3] 
titles=["1st derivative_xy","2nd derivative_xy","3rd derivative_xy"]
fig,ax=plt.subplots(1,3,figsize=(10,10))
for i in range (3):
    ax[i].imshow(cv.Sobel(img,-1,h[i],h[i],ksize=5),cmap='gray')
    ax[i].set_xticks([]),ax[i].set_yticks([])
    ax[i].set_title(titles[i])
plt.show()
```

```python
#Laplace kernal.Adds 2nd derivative wrt x and y 
fig,ax=plt.subplots(1,2,figsize=(9,9))
ax[0].imshow(cv.Laplacian(img,-1,ksize=5),cmap='gray')
ax[1].imshow((cv.Sobel(img,-1,0,2,ksize=5)+cv.Sobel(img,cv.CV_8U,2,0,ksize=5)),cmap='gray')
ax[0].set_xticks([]),ax[0].set_yticks([])
ax[0].set_title("Using Laplace filter")
ax[1].set_xticks([]),ax[1].set_yticks([])
ax[1].set_title("Using Sobel filter wrt x and y and adding them")
plt.tight_layout()
plt.show()
```

```python
#Seamless cloning
img_eye=cv.imread('Downloads\eye.jpg')
img_pe=cv.imread('Downloads\peper.png')
dst= cv.cvtColor(img_eye, cv.COLOR_BGR2RGB)
src= cv.cvtColor(img_pe, cv.COLOR_RGB2BGR)
src_mask = np.zeros(src.shape, src.dtype)
poly = np.array([[0,0],[20,50],[100,120],[200,250],[200,280],[500,200]], np.int32)
cv.fillPoly(src_mask, [poly],(255,255,255),cv.LINE_8)
center=(1500,600)
output = cv.seamlessClone(src, dst, src_mask, center, cv.NORMAL_CLONE)
plt.imshow(output)
plt.xticks([]),plt.yticks([])
plt.title("Seamless Cloning")
plt.show()
```
