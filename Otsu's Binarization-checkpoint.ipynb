```python
import numpy as np 
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
leaf=cv.imread('Downloads\leaf.jpg',0)
plt.imshow(leaf,'gray')
print("Shape:",leaf.shape)
```

```python
plt.hist(leaf.ravel(),256)
plt.show()
```

```python
ret,th1=cv.threshold(leaf,127,255,cv.THRESH_BINARY)#binary thresholding directly 
ret,th2 = cv.threshold(leaf,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)#binary thresholding+otsu's thresholding
leaf_blur=cv.GaussianBlur(leaf,(5,5),0)#remove noise in the image 
ret,th3=cv.threshold(leaf_blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)##binary thresholding+otsu's thresholding on blur image
plot1=[leaf,th1,leaf,th2,leaf_blur,th3]
titles1=['Histrogram','Binary Threshld','Histrogram','Binary+Otsu Threshold','Histrogram_blur','Binary+Otsu Threshold_blur']
for i in range(3):
    plt.subplot(3,2,(i*2)+1),plt.hist(plot1[i*2].ravel(),256)
    plt.title(titles1[i*2]),plt.xticks([]), plt.yticks([])
    plt.subplot(3,2,(i*2)+2),plt.imshow(plot1[(i*2)+1],'gray')
    plt.title(titles1[(i*2)+1]),plt.xticks([]), plt.yticks([])
    plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
ret,th1=cv.threshold(leaf,127,255,cv.THRESH_BINARY_INV)#binary inverse thresholding 
ret,th2 = cv.threshold(leaf,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)
leaf_blur=cv.GaussianBlur(leaf,(5,5),0)
ret,th3=cv.threshold(leaf_blur,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)
plot1=[leaf,th1,leaf,th2,leaf_blur,th3]
titles1=['Histrogram','B_inv Threshld','Histrogram','B_inv+Otsu Threshold','Histrogram_blur','B_inv+Otsu Threshold_blur']
for i in range(3):
    plt.subplot(3,2,(i*2)+1),plt.hist(plot1[i*2].ravel(),256)
    plt.title(titles1[i*2]),plt.xticks([]), plt.yticks([])
    plt.subplot(3,2,(i*2)+2),plt.imshow(plot1[(i*2)+1],'gray')
    plt.title(titles1[(i*2)+1]),plt.xticks([]), plt.yticks([])
    plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
img=cv.imread('Downloads\lenagray.bmp',0)
plt.imshow(img,'gray')
print("Shape:",img.shape)
```

```python
plt.hist(img.ravel(),256)
plt.ylim([0,6000])
plt.show()
```

```python
#according to the histrograms we cannot identify 2 clear seperate peaks. Rather they are distributed all most equally 
#(multimodel image). In order to perform Otsu' binarization it has to be a bimodel images (most pixels in the image are 
#distributed over 2 values so Otsu's method can determine an optimal global threshold value in between those peaks.
ret,th1=cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,th2 = cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
#perform gassianBlur on the image to remove noise. 
img_blur=cv.GaussianBlur(img,(5,5),0)
ret,th3=cv.threshold(img_blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
plot2=[img,th1,img,th2,img_blur,th3]
titles2=['Histrogram','Binary Threshld','Histrogram','Binary+Otsu Threshold','Histrogram_blur','Binary+Otsu Threshold_blur']
for i in range(3):
    plt.subplot(3,2,(i*2)+1),plt.hist(plot2[i*2].ravel(),256)
    plt.title(titles2[i*2]),plt.xticks([]), plt.yticks([])
    plt.subplot(3,2,(i*2)+2),plt.imshow(plot2[(i*2)+1],'gray')
    plt.title(titles2[(i*2)+1]),plt.xticks([]), plt.yticks([])
    plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
#as you can see results are not visible as before
```
