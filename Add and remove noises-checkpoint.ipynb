```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt 
```

```python
img=cv.imread("Downloads/peper.png",0)
```

```python
#add gaussian noise to the img for different zigma values 
zigma=[10,20,30]
img_nois=[]
for j in zigma:  
    gauss = np.random.normal(0,j,img.size)
    gauss = gauss.reshape(img.shape[0],img.shape[1])
    no=img+gauss
    no=np.clip(no,0.0,255.0)
    img_nois.append(no)
titles=["Img+noise 10","Img+noise 20","Img+noise 30"]
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(img_nois[i],cmap='gray',vmin=0,vmax=255)
    plt.xticks([]),plt.yticks([])
    plt.title(titles[i])
plt.show()
```

```python
#histograms of original img and nois images 
his=[]
his.append(img)
for y in img_nois:
    his.append(y)
ti=["Histo of img","Histo 10-noise img","Histo 20-noise img","Histo 30-noise img"]
fig,ax=plt.subplots(2,2,figsize=(7,7))
ax[0,0].hist(his[0].ravel(),256,[0,256])
ax[0,0].set_title(ti[0])
ax[0,1].hist(his[1].ravel(),256,[0,256])
ax[0,1].set_title(ti[1])
ax[1,0].hist(his[2].ravel(),256,[0,256])
ax[1,0].set_title(ti[2])
ax[1,1].hist(his[3].ravel(),256,[0,256])
ax[1,1].set_title(ti[3])
plt.tight_layout()    
```

```python
#remove gaussian noise by gaussian filtering 
fig,ax=plt.subplots(3,3,figsize=(10,10))
titles1=["Blured noise-10 img 1px","Blured noise-20 img 1px","Blured noise-30 img 1px"]
titles2=["Blured noise-10 img 3px","Blured noise-20 img 3px","Blured noise-30 img 3px"]
for k in range(3):
    ax[k,0].imshow(img_nois[k],cmap='gray')
    ax[k,0].set_xticks([]),ax[k,0].set_yticks([])
    ax[k,0].set_title(titles[k])
    ax[k,1].imshow(cv.GaussianBlur(img_nois[k],(1,1),0),cmap='gray')#filtering kernal size 1x1
    ax[k,1].set_xticks([]),ax[k,1].set_yticks([])
    ax[k,1].set_title(titles1[k])
    ax[k,2].imshow(cv.GaussianBlur(img_nois[k],(3,3),0),cmap='gray')#filtering kernal size 3x3
    ax[k,2].set_xticks([]),ax[k,2].set_yticks([])
    ax[k,2].set_title(titles2[k])
```

```python
randt=np.random.rand(img.shape[0],img.shape[1])*255
randt.astype('uint8')
Noise=img
Noise[randt<=25]=0
Noise[randt>=230]=255#add salt and peper noise to the img 
med=cv.medianBlur(Noise,5)#remove that noise using medain blur
fig,ax=plt.subplots(1,2,figsize=(7,7))
ax[0].imshow(Noise,cmap='gray')
ax[1].imshow(med,cmap='gray')
ax[0].set_xticks([]),ax[0].set_yticks([])
ax[0].set_title("Salt and peper noise img")
ax[1].set_xticks([]),ax[1].set_yticks([])
ax[1].set_title("Median filtered noise img")
plt.show()
```

```python
#adding speckle noise to the img
spec_gauss = np.random.normal(0,0.7,img.size)
spec_gauss = spec_gauss.reshape(img.shape[0],img.shape[1])
spec=img+img*spec_gauss
spec=np.clip(spec,0.0,255.0)
plt.imshow(spec,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.title("Speckle noise added img")
plt.show()
```

```python
from scipy.ndimage.filters import uniform_filter
from scipy.ndimage.measurements import variance
```

```python
#implimenting lee filter and using it to remove speckle noise 
img_output=[]
zl=[3,15,21]
for z in zl:
    kernal_avg=np.ones((z,z),np.float32)/(z**2)
    spec_mean = cv.filter2D(spec,-1,kernal_avg)
    spec_sqr_mean = cv.filter2D(spec**2,-1,kernal_avg)
    spec_variance = spec_sqr_mean - spec_mean**2
    overall_variance = variance(img)
    weights = spec_variance / (spec_variance + overall_variance)
    output=spec_mean + weights * (spec - spec_mean)
    output=np.clip(output,0.0,255.0)
    img_output.append(output)
```

```python
#but as you can see it is not effective on edge area
fig,ax=plt.subplots(1,3,figsize=(10,10))
ax[0].imshow(img_output[0],cmap='gray')
ax[0].set_xticks([]),ax[0].set_yticks([])
ax[0].set_title("Window size 3x3 lee filter")
ax[1].imshow(img_output[1],cmap='gray')
ax[1].set_xticks([]),ax[1].set_yticks([])
ax[1].set_title("Window size 15x15 lee filter")
ax[2].imshow(img_output[2],cmap='gray')
ax[2].set_xticks([]),ax[2].set_yticks([])
ax[2].set_title("Window size 21x21 lee filter")
plt.show()
```
