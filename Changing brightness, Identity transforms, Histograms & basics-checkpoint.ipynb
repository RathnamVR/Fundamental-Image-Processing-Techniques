```python
#additive nature of computer display 
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt 
img=cv.imread('Downloads\peper.png')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
img[:,:,0:2]=0#0 outing red and green channels. Output will be still a 3 channel array.Red & green channels will be 
#completely black.
fig,axis=plt.subplots(figsize=(5,5))
plt.imshow(img)
```

```python
img=cv.imread('Downloads\peper.png',0)
print("Size:",img.size)
print("Type:",img.dtype)
```

```python
img_a=img+100#adding 100 to each pixel in the image 
img_b=cv.add(img,100)#adding 100 using cv function add
plots=[img,img_a,img_b]
titles=['Original','Direct add 100','Add 100 through cv.add']
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(plots[i],cmap='gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.tight_layout()
plt.show()
#When adding 100 directly some pixels go beyond 255 and those pixels will be treated ignoring the overflow part 
#When adding 100 through cv the maximum a pixel can reach is 255 so there is no overflow 
```

```python
def brightness(image,shift):
    h=image.shape[0]
    w=image.shape[1]
    result=np.zeros((h,w),image.dtype)
    for i in range(h):
        for j in range(w):
            result[i,j]=min(image[i,j]+100,255)
    return(result)
```

```python
img_c=brightness(img,100)
plt.imshow(img_c,cmap='gray',vmin=0,vmax=255)#But inefficient 
```

```python
#Itensity transformations 
#Identity transformation 
x=np.arange(0,256).astype('uint8')
y=x#Identity function 
fig,ax=plt.subplots()
ax.plot(y)
ax.set_aspect(1)
ax.set_title("Transformation")
plt.show()
```

```python
img_t=cv.LUT(img,y)#transformation using cv.LUT
img_t2=y[img]#transformation using Numpy
plots=[img,img_t,img_t2]
titles=['Original','cv.LUT transform','Numpy transform']
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(plots[i],cmap='gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
#negative transformation or inevrse identity transformation
y_inv=-x+255
fig,ax=plt.subplots()
ax.plot(y_inv)
ax.set_aspect(1)
ax.set_title("Negative transform/Identity inverse transform")
plt.show()
```

```python
img_tinv=cv.LUT(img,y_inv)
img_t2inv=y_inv[img]
plots=[img,img_tinv,img_t2inv]
titles=['Original','cv.LUT transform','Numpy transform']
for j in range(3):
    plt.subplot(1,3,j+1),plt.imshow(plots[j],cmap='gray',vmin=0,vmax=255)
    plt.title(titles[j])
    plt.xticks([]),plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
#Intensity windowing 
x1=np.arange(0,101).astype('uint8')
y1=(x1//2)
x2=np.arange(101,151).astype('uint8')
y2=3*x2-250
x3=np.arange(151,256).astype('uint8')
y3=(x3//(21/11))+122
y3=y3.astype('uint8')
trans=np.concatenate((y1,y2),axis=0).astype('uint8')
trans=np.concatenate((trans,y3),axis=0).astype('uint8')
fig,ax=plt.subplots(figsize=(5,5))
ax.plot(trans)
ax.set_title("Intensity window transformation")
ax.set_aspect(1)
plt.show()
```

```python
img_inw=cv.LUT(img,trans)
img_inw2=trans[img]
plots=[img,img_inw,img_inw2]
titles=['Original','cv.LUT transform','Numpy transform']
for k in range(3):
    plt.subplot(1,3,k+1),plt.imshow(plots[k],cmap='gray',vmin=0,vmax=255)
    plt.title(titles[k])
    plt.xticks([]),plt.yticks([])
plt.tight_layout()
plt.show()
```
