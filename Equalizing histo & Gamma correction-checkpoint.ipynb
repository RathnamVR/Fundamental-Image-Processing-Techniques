```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_gray=cv.imread('Downloads\peper.png',0)
```

```python
histro,bins=np.histogram(img_gray.ravel(),256,[0,256])#original histrogram 
cum_sum=histro.cumsum()
cdf_norm=cum_sum*histro.max()/cum_sum.max()#normalized cdf. To match the histogram scaling we have to normalize.
plt.hist(img_gray.ravel(),256,[0,256],color='r')
plt.plot(cdf_norm,color='b')
plt.legend(('cdf_norm','histogram','cdf'),loc='upper left')
plt.xlim=([0,256])
plt.show()
```

```python
eq_out=np.around(cum_sum*255/cum_sum.max())#equalizer output for each input pixel value. Index value correspond to input pixel 
#value
plt.plot(eq_out)
print(eq_out)
```

```python
img_eq=cv.equalizeHist(img_gray)#cv function for equalizer 
histro2,bins=np.histogram(img_eq.ravel(),256,[0,256])#equalized image's hostogram 
cum_sum2=histro2.cumsum()
cdf_norm=cum_sum2*histro2.max()/cum_sum2.max()#normalized cdf 
plt.hist(img_eq.ravel(),256,[0,256],color='r')
plt.plot(cdf_norm,color='b')
plt.legend(('cdf_norm','histogram'),loc='upper left')
plt.xlim=([0,256])
plt.show()
```

```python
images=[img_gray,img_eq]
titles=['Original img','Equalized img']
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],cmap='gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img=cv.imread('Downloads\peper.png')
img=img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
table=np.array([(i/255.0)**(1/2)*255.0 for i in np.arange(0,256)]).astype('uint8')#gamma=1/2
img_gam=cv.LUT(img,table)#gamma correction on image.img_gam[i]<---table[img[i]]
```

```python
images=[img,img_gam]
titles=['Original img','Gamma<1 img']
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],cmap='gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,12))
c=['b','g','r']
for l in range(3):
    hist=cv.calcHist([img],[l],None,[256],[0,256])#color histograms of input image 
    ax[0].plot(hist,color=c[l])
    ax[0].set_aspect(0.01)
    hist_=cv.calcHist([img_gam],[l],None,[256],[0,256])#color histograms of output image 
    ax[1].plot(hist_,color=c[l])
    ax[1].set_aspect(0.01)
plt.tight_layout()
plt.show()  
```

```python
table2=np.array([(i/255.0)**(2)*255.0 for i in np.arange(0,256)]).astype('uint8')#gamma=2
img_gam2=cv.LUT(img,table2)
```

```python
images=[img,img_gam2]
titles=['Original img','Gamma>1 img']
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],cmap='gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,12))
c=['b','g','r']
for l in range(3):
    hist1=cv.calcHist([img],[l],None,[256],[0,256])
    ax[0].plot(hist1,color=c[l])
    ax[0].set_aspect(0.01)
    hist_1=cv.calcHist([img_gam2],[l],None,[256],[0,256])
    ax[1].plot(hist_1,color=c[l])
    ax[1].set_aspect(0.005)
plt.tight_layout()
plt.show()  
```

```python
#here the input image is a colored one. We can do gamma correction to 3 channels seperatedly using a 3 channel table. Earlier
#all 3 channels were gamma corrected by same table  
table_r=np.array([(i/255.0)**(1/4)*255.0 for i in np.arange(0,256)]).astype('uint8')
table_b=np.array([(i/255.0)**(1/6)*255.0 for i in np.arange(0,256)]).astype('uint8')
table_g=np.array([(i/255.0)**(1/3)*255.0 for i in np.arange(0,256)]).astype('uint8')
tabel=np.dstack((table_r,table_b,table_g))
img_gam3=cv.LUT(img,tabel)
plt.imshow(img_gam3)
```
