```python
th=[50,100,127,200,230]
truncthresholdimage=[]
for i in th:
    ret,thresh = cv.threshold(img_lena,i,255,cv.THRESH_TRUNC)#pixels with values more than threshold 
    truncthresholdimage.append(thresh)
for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(truncthresholdimage[i],'gray',vmin=0,vmax=255)
    plt.title(th[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
