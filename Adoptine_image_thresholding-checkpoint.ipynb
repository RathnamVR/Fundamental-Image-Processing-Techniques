```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt 
img=cv.imread('Downloads\lenagray.bmp',0)
plt.imshow(img,'gray')
print("Shape:",img.shape)
```

```python
img=cv.medianBlur(img,5)#blurs the image a little
plt.imshow(img,'gray')
```

```python
ret,thresh=cv.threshold(img,127,255,cv.THRESH_BINARY)#binary threshold the image
thresh_mean=cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
                                 cv.THRESH_BINARY,11,2)#do binary theshold in adaptive
#threshold mean methord. First integer of last 2 is neighbourhood size. Last integer is constant to be subtracted 
thresh_gaussian=cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
                                     cv.THRESH_BINARY,11,2)#do binary theshold in 
#adaptive threshold gaussian methord.
#In these last 2 methords threshold value is calculated seperately for each pixel depending on there neighbouring pixels. 
Title=['Global Binary threshold','Adaptive mean binary threshold','Adaptive gaussian binary threshold']
plot=[thresh,thresh_mean,thresh_gaussian]
for i in range(3):
    plt.subplot(2,2,i+1),plt.imshow(plot[i],'gray')
    plt.title(Title[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
ret,thresh=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)#inverse binary threshold the image
thresh_mean=cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
                                 cv.THRESH_BINARY_INV,11,2)
thresh_gaussian=cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
                                     cv.THRESH_BINARY_INV,11,2)
Title=['Global b_inv threshold','Adaptive mean b_inv threshold','Adaptive gaussian b_inv threshold']
plot=[thresh,thresh_mean,thresh_gaussian]
for i in range(3):
    plt.subplot(2,2,i+1),plt.imshow(plot[i],'gray')
    plt.title(Title[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
